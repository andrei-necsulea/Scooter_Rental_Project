```python
from xgboost import XGBRegressor, plot_importance
```

```python
scooter_rental_data['hour'] = scooter_rental_data['datetime'].dt.hour
scooter_rental_data['weekday'] = scooter_rental_data['datetime'].dt.weekday
scooter_rental_data['month'] = scooter_rental_data['datetime'].dt.month
scooter_rental_data['year'] = scooter_rental_data['datetime'].dt.year
scooter_rental_data['day'] = scooter_rental_data['datetime'].dt.day

scooter_rental_data['hour'] = pd.to_numeric(scooter_rental_data['hour'], errors='coerce')
scooter_rental_data['hour_sin'] = np.sin(2*np.pi * scooter_rental_data['hour'] / 24)
scooter_rental_data['hour_cos'] = np.cos(2*np.pi * scooter_rental_data['hour'] / 24)

#scooter_rental_data = scooter_rental_data.drop(columns=['datetime'])
scooter_rental_data['datetime'] = pd.to_numeric(scooter_rental_data['datetime'])

scooter_rental_data['is_weekend'] = scooter_rental_data['weekday'].isin([5, 6]).astype(int)
scooter_rental_data['is_day_off'] = ((scooter_rental_data['is_weekend'] == 1) | (scooter_rental_data['holiday'] == 1)).astype(int)

scooter_COPY = scooter_rental_data.copy()

#Am uitat de nan-uri, le scot aici
scooter_rental_data = scooter_rental_data.dropna(axis=0, how='any')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numeric_cols = ['temp','hour', 'year', 'humidity', 'windspeed', 'atemp', 'hour_sin', 'hour_cos', 'datetime']
cat_cols = [
    'holiday','workingday', 'is_weekend', 'is_day_off',
    'season_Autumn','season_Winter','season_Summer', 'weekday', 'weather_Clear_Few_clouds_Partly_cloudy', 'weather_Mist_Cloudy_Mist_Broken_clouds_Mist_Few_clouds_Mist', 'weather_Light_Snow_Light_Rain_Thunderstorm_Scattered_clouds_Light_Rain_Scattered_clouds',
     'weather_Heavy_Rain_Hail_Thunderstorm_Mist_Snow_Fog', 'season_Spring'
]

scooter_rental_data_scaled = scaler.fit_transform(scooter_rental_data[numeric_cols])

new_df = pd.DataFrame(scooter_rental_data_scaled, columns=numeric_cols, index=scooter_rental_data.index)

X = pd.concat([new_df, scooter_rental_data[cat_cols]], axis=1)

y = scooter_rental_data['count']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor
from lightgbm import LGBMRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, LassoCV, ElasticNetCV, RidgeCV
from sklearn.svm import SVR, LinearSVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
EPS = 1e-9

def mape(y_true, y_pred, ignore_zeros=True):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    if ignore_zeros:
        mask = y_true != 0
        if not np.any(mask):
            return np.nan
        y_true = y_true[mask]; y_pred = y_pred[mask]
    return 100.0 * np.mean(np.abs((y_true - y_pred) / (y_true + EPS)))

def safe_mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    return 100.0 * np.mean(np.abs(y_true - y_pred) / (denom + EPS))

def metrics(y_true, y_pred, model):
    model_name = model.__class__.__name__
    mse  = mean_squared_error(y_true, y_pred)
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    mape_pct  = mape(y_true, y_pred, ignore_zeros=True)
    smape_pct = safe_mape(y_true, y_pred)
    print(f"\n=== {model_name} ===")
    print(f"MSE:   {mse:.6f}")
    print(f"RMSE:  {rmse:.6f}")
    print(f"MAE:   {mae:.6f}")
    print(f"MAPE:  {mape_pct:.2f}%")
    print(f"SMAPE: {smape_pct:.2f}%")
    print(f"R2:    {r2:.6f}")
    return {"Model": model_name, "MSE": mse, "RMSE": rmse, "MAE": mae,
            "MAPE%": mape_pct, "SMAPE%": smape_pct, "R2": r2}

param_dict = {
    # ======================
    # MODELE LINIARE
    # ======================
    "LinearRegression": {},  # nu are hiperparametri majori

    "Lasso": {
        "alpha": 120,  # L1 penalty strength
        "max_iter": 10000
    },

    "Ridge": {
        "alpha": 120,  # L2 penalty strength
        "max_iter": 10000
    },

    "ElasticNet": {
        "alpha": 240,
        "l1_ratio": 0.9,
        "max_iter": 10000
    },

    # ======================
    # SUPPORT VECTOR MACHINES
    # ======================
    "SVR": {
        "C": 50,
        "epsilon": 0.2,
        "kernel": "rbf"
    },

    "LinearSVR": {
        "C": 10,
        "epsilon": 0.2,
        "max_iter": 10000
    },

    # ======================
    # NEAREST NEIGHBORS
    # ======================
    "KNN": {
        "n_neighbors": 100,
        "weights": "uniform",
        "p": 2 # 1=manhattan, 2=euclidian
    },

    # ======================
    # ARBORI / ENSEMBLE
    # ======================
    "DecisionTree": {
        "max_depth": None,
        "min_samples_split": 2,
        "min_samples_leaf": 2
    },

    "RandomForest": {
        'n_estimators':500,          # 600â€“1200
        'max_depth':None,
        'min_samples_split':5,      # 8â€“20
        'min_samples_leaf':2,        # 2â€“6
        'max_features':None,       # <<< NU lÄƒsa None; âˆšN creÈ™te diversitatea
        'bootstrap':True,
        'criterion':'absolute_error', # Ã®ncearcÄƒ È™i 'poisson' dacÄƒ y>=0
        'oob_score':True,
        'n_jobs':-1,
    },

    "ExtraTrees": {
        "n_estimators": 500,
        "max_depth": None,
        "min_samples_split": 2,
        "min_samples_leaf": 2,
        "max_features": None,
        'bootstrap': True,
        'criterion': 'absolute_error',
        'oob_score': True,
        'n_jobs': -1,
    },

    # ======================
    # XGBOOST
    # ======================
    "XGB": {
        "n_estimators": 100,
        "learning_rate": 0.1,
        "max_depth": None,
        "min_child_weight": 5,
        "subsample": 0.9,
        "colsample_bytree": 0.9,
        "reg_alpha": 0.5,
        "reg_lambda": 2.0,
        "gamma": 0.3
    },

    "LassoCV": {
        "alphas": 120,  # range de cÄƒutare pentru alpha
        "cv": 10,                    # numÄƒr de folduri
        "max_iter": 10000,
        "selection": "cyclic"
    },

    "ElasticNetCV": {
        "l1_ratio": .9,  # testÄƒm mai multe rapoarte L1/L2
        "alphas": 240,
        "cv": 10,
        "max_iter": 10000,
        "selection": "cyclic"
    },

    "RidgeCV": {
        "alphas": np.logspace(-3, 2, 50),   # array direct, nu [array]
        "cv": 10,
        "scoring": "neg_mean_absolute_error"
    },

    "GradientBoosting": {
        "n_estimators": 800,
        "learning_rate": 0.1,
        "max_depth": 7,
        "min_samples_split": 10,
        "min_samples_leaf": 4,
        "subsample": 1.0,
        "max_features": None
    },

    "LGBM" : {
        "n_estimators": 500,
        "learning_rate": 0.1,
        "num_leaves": 127,
        "max_depth": -1,
        "subsample": 1.0,
        "colsample_bytree": 1.0,
        "reg_alpha": 0.5,
        "reg_lambda": 2.0
    }
}

lr = LinearRegression()
lr.set_params(**param_dict["LinearRegression"])
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
metrics(y_test, y_pred_lr, lr)

lasso = Lasso()
lasso.set_params(**param_dict["Lasso"])
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
metrics(y_test, y_pred_lasso, lasso)

Ridge = Ridge()
Ridge.set_params(**param_dict["Ridge"])
Ridge.fit(X_train, y_train)
y_pred_ridge = Ridge.predict(X_test)
metrics(y_test, y_pred_ridge, Ridge)

en = ElasticNet()
en.set_params(**param_dict["ElasticNet"])
en.fit(X_train, y_train)
y_pred_en = en.predict(X_test)
metrics(y_test, y_pred_en, en)

LassoCV = LassoCV()
LassoCV.set_params(**param_dict["LassoCV"])
LassoCV.fit(X_train, y_train)
y_pred_lasso_cv = LassoCV.predict(X_test)
metrics(y_test, y_pred_lasso_cv, LassoCV)

encv = ElasticNetCV()
encv.set_params(**param_dict["ElasticNetCV"])
encv.fit(X_train, y_train)
y_pred_en_cv = encv.predict(X_test)
metrics(y_test, y_pred_en_cv, encv)

RidgeCV = RidgeCV()
RidgeCV.set_params(**param_dict["RidgeCV"])
RidgeCV.fit(X_train, y_train)
y_pred_ridge_cv = RidgeCV.predict(X_test)
metrics(y_test, y_pred_ridge_cv, RidgeCV)

SVR = SVR()
SVR.set_params(**param_dict["SVR"])
SVR.fit(X_train, y_train)
y_pred_svr = SVR.predict(X_test)
metrics(y_test, y_pred_svr, SVR)

lsvr = LinearSVR()
lsvr.set_params(**param_dict["LinearSVR"])
lsvr.fit(X_train, y_train)
y_pred_lsvr = lsvr.predict(X_test)
metrics(y_test, y_pred_lsvr, lsvr)

rf = RandomForestRegressor()
rf.set_params(**param_dict["RandomForest"])
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
metrics(y_test, y_pred_rf, rf)

etr = ExtraTreesRegressor()
etr.set_params(**param_dict["ExtraTrees"])
etr.fit(X_train, y_train)
y_pred_etr = etr.predict(X_test)
metrics(y_test, y_pred_etr, etr)

DecisionTreeRegressor = DecisionTreeRegressor()
DecisionTreeRegressor.set_params(**param_dict["DecisionTree"])
DecisionTreeRegressor.fit(X_train, y_train)
y_pred_dt = DecisionTreeRegressor.predict(X_test)
metrics(y_test, y_pred_dt, DecisionTreeRegressor)

knn = KNeighborsRegressor()
knn.set_params(**param_dict["KNN"])
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
metrics(y_test, y_pred_knn, knn)

xgb = XGBRegressor()
xgb.set_params(**param_dict["XGB"])
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
metrics(y_test, y_pred_xgb, xgb)

gbr = GradientBoostingRegressor()
gbr.set_params(**param_dict["GradientBoosting"])
gbr.fit(X_train, y_train)
y_pred_gbr = gbr.predict(X_test)
metrics(y_test, y_pred_gbr, gbr)

lgbm = LGBMRegressor()
lgbm.set_params(**param_dict["LGBM"])
lgbm.fit(X_train, y_train)
y_pred_lgbm = lgbm.predict(X_test)
metrics(y_test, y_pred_lgbm, lgbm)

rf_smape = safe_mape(y_test, y_pred_rf)
etr_smape = safe_mape(y_test, y_pred_etr)
knn_smape = safe_mape(y_test, y_pred_knn)
lsvr_smape = safe_mape(y_test, y_pred_lsvr)
ridge_smape = safe_mape(y_test, y_pred_ridge)
lr_smape = safe_mape(y_test, y_pred_lr)
elastic_net_smape = safe_mape(y_test, y_pred_etr)
lasso_smape = safe_mape(y_test, y_pred_lasso)
lasso_cv_smape = safe_mape(y_test, y_pred_lasso_cv)
ridge_cv_smape = safe_mape(y_test, y_pred_ridge_cv)
svr_smape = safe_mape(y_test, y_pred_svr)
dt_smape = safe_mape(y_test, y_pred_dt)
elasticnet_cv_smape = safe_mape(y_test, y_pred_en_cv)
xgb_smape = safe_mape(y_test, y_pred_xgb)
gbr_smape = safe_mape(y_test, y_pred_gbr)
lgbm_smape = safe_mape(y_test, y_pred_lgbm)

rf_mape = mape(y_test, y_pred_rf, ignore_zeros=True)
etr_mape = mape(y_test, y_pred_etr, ignore_zeros=True)
lr_mape = mape(y_test, y_pred_lr, ignore_zeros=True)
ridge_mape = mape(y_test, y_pred_ridge_cv, ignore_zeros=True)
lsvr_mape = mape(y_test, y_pred_lsvr, ignore_zeros=True)
elasticnet_mape = mape(y_test, y_pred_en_cv, ignore_zeros=True)
lasso_mape = mape(y_test, y_pred_lasso, ignore_zeros=True)
lasso_cv_mape = mape(y_test, y_pred_lasso_cv, ignore_zeros=True)
knn_mape = mape(y_test, y_pred_knn, ignore_zeros=True)
ridge_cv_mape = mape(y_test, y_pred_ridge_cv, ignore_zeros=True)
lsvr_cv_mape = mape(y_test, y_pred_lsvr, ignore_zeros=True)
elasticnet_cv_mape = mape(y_test, y_pred_en_cv, ignore_zeros=True)
dt_mape = mape(y_test, y_pred_dt, ignore_zeros=True)
svr_mape = mape(y_test, y_pred_svr, ignore_zeros=True)
xgb_mape = mape(y_test, y_pred_xgb, ignore_zeros=True)
gbr_mape = mape(y_test, y_pred_gbr, ignore_zeros=True)
lgbm_mape = mape(y_test, y_pred_lgbm, ignore_zeros=True)

smape_dict = {
    'RandomForestRegressor': rf_smape,
    'ExtraTreesRegressor': etr_smape,
    'LinearRegression': lr_smape,
    'Lasso': lasso_smape,
    'Ridge': ridge_smape,
    'ElasticNet': elastic_net_smape,
    'LassoCV': lasso_cv_smape,
    'ElasticNetCV': elasticnet_cv_smape,
    'RidgeCV': ridge_cv_smape,
    'KNeighborsRegressor': knn_smape,
    'LinearSVR': lsvr_smape,
    'DecisionTreeRegressor': dt_smape,
    'SVR': svr_smape,
    'XGBoostRegressor': xgb_smape,
    'GBRTRegressor': gbr_smape,
    'LGBMRegressor': lgbm_smape,
}

mape_dict = {
    'RandomForestRegressor': rf_mape,
    'ExtraTreesRegressor': etr_mape,
    'LinearRegression': lr_mape,
    'Lasso': lasso_mape,
    'Ridge': ridge_mape,
    'ElasticNet': elasticnet_mape,
    'LassoCV': lasso_cv_mape,
    'ElasticNetCV': elasticnet_cv_mape,
    'RidgeCV': ridge_cv_mape,
    'KNeighborsRegressor': knn_mape,
    'LinearSVR': lsvr_mape,
    'DecisionTreeRegressor': dt_mape,
    'SVR': svr_mape,
    'XGBoostRegressor': xgb_mape,
    'GBRTRegressor': gbr_mape,
    'LGBMRegressor': lgbm_mape,
}

print(f"\nBest model by MAPE is {min(mape_dict, key=mape_dict.get)} with MAPE = {min(mape_dict.values()):.2f}%")
print(f"\nBest model by SMAPE is {min(smape_dict, key=smape_dict.get)} with SMAPE = {min(smape_dict.values()):.2f}%")
```

```python
from sklearn.model_selection import TimeSeriesSplit, cross_validate
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import make_scorer, r2_score

#pregatire X, y numerice
X = scooter_rental_data.drop(columns=['count'], errors='ignore')
X = X.select_dtypes(include=['number']).copy()
y = scooter_rental_data['count'].astype(float).values

def safe_mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = y_true > 0
    if not np.any(mask):
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100.0

def smape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    mask = denom != 0
    if not np.any(mask):
        return np.nan
    return np.mean(np.abs(y_true[mask] - y_pred[mask]) / denom[mask]) * 100.0

mape_scorer  = make_scorer(safe_mape, greater_is_better=False)   # sklearn va NEGAt rezultatul (pt cÄƒ e eroare)
smape_scorer = make_scorer(smape,      greater_is_better=False)

tscv = TimeSeriesSplit(n_splits=10)

etr_dict = {
        "n_estimators": 500,
        "max_depth": None,
        "min_samples_split": 2,
        "min_samples_leaf": 2,
        "max_features": None,
        'bootstrap': True,
        'criterion': 'absolute_error',
        'oob_score': True,
        'n_jobs': -1,
}

model = ExtraTreesRegressor()
model.set_params(**etr_dict)

scores = cross_validate(
    model, X, y, cv=tscv, n_jobs=-1,
    scoring={'R2': 'r2', 'MAPE': mape_scorer, 'SMAPE': smape_scorer},
    return_train_score=False
)

r2_mean    = np.nanmean(scores['test_R2'])
mape_mean  = -np.nanmean(scores['test_MAPE'])
smape_mean = -np.nanmean(scores['test_SMAPE'])

print("R2 (mean CV):   ", round(r2_mean, 4))
print("MAPE% (mean CV):", round(mape_mean, 2))
print("SMAPE% (mean CV):", round(smape_mean, 2))
print("Fold R2:", np.round(scores['test_R2'], 4))
print("Fold MAPE%:", np.round(-scores['test_MAPE'], 2))
print("Fold SMAPE%:", np.round(-scores['test_SMAPE'], 2))
```
